#### Exercício 1:

Use o método de Euler implementado nas aulas anteriores para estimar os valores
sucessivos da solução do valor inicial

$\begin{cases} y' = -10y \\ y(0) = 1 \end{cases}$, com $t \in [0,5]$


a) $h=0.01$

```python
def euler_1a_ordem(f, y0, h, t_range):
    
    """
    Resolve o problema de valor inicial com o método de Euler de 1ª ordem.
    
    :param f: função lambda(t,y) de y' do problema
    :param y0: ponto inicial do problema
    :param h: passo escolhido
    :param t_range: intervalo em t ([lower, upper])
    
    :returns (t,y): estimação de y nos valores em t
    """
        
    y_ant = y0[1]
    t_ant = y0[0]
    
    t_est = [t_ant]
    y_est = [y_ant]
    
    while t_ant < t_range[1]:
        
        # passo de Taylor
        t = t_ant + h
        y = y_ant + f(t_ant, y_ant)*h
        
        y_est.append(y)
        t_est.append(t)
        
        # atualização dos termos
        y_ant = y
        t_ant = t
    
    return t_est, y_est
```

```python
def plot_aprox(*plots):
    
    """
    Plota uma ou mais aproximações pelos métodos escolhidos.
    
    :param metodo:
    :param f:
    :param y0:
    :param h: 
    :param t_range:
    :param label:
    """

    y = []
    
    for plot in plots:
        
        metodo = plot[0]
        f = plot[1]
        y0 = plot[2]
        h = plot[3]
        t_range = plot[4]
        label = plot[5]
    
        t_est, y_est = metodo(f, y0, h, t_range)

        plt.plot(t_est, y_est, label=label)
        y.append(y_est)
        
    plt.legend()
    plt.show()

    return y
```

```python
f = lambda t, y: -10*y
y0 = (0,1)
t_range = [0,5]
#t = np.linspace(0, 5, 10)
```

```python
h = 0.01

plot_aprox([euler_1a_ordem, f, y0, h, t_range, 'Euler aprox'])
```

### [Método de Euler implícito](https://www.if.ufrgs.br/tex/metcomp1/aulas/euler/node5.html)

Semelhante ao método de Euler original, usamos também na solução de equações diferenciais ordinárias da forma:

$\begin{cases} y′ = f(y,t) \\ y(t_{0}) = y_0 \end{cases}$

Mas, agora, supomos que é conhecida $f(Y(t+h))$, e a utilizamos na aproximação de $Y(t+h)$:

$\begin{cases} Y(t+h) = Y(t) + f(Y(t+h), t+h)h \\ f(Y(t), t) = y' \end{cases}$

Ou seja, olhamos a derivada no próprio ponto para descobrir seu valor, ao invés de usar a derivada no ponto anterior.

**OBS:** A vantagem deste método em relação ao original é sua estabilidade.

#### Exercício 2:

Implemente o método de Euler implícito. Refaça o experimento do exercício (1) com o método de Euler implícito para observar a estabilidade do método.


**R:** Neste caso, vamos resolver analiticamente a aproximação, dado o problema de valor inicial:

$\begin{cases} y' = -10y \\ y(0) = 1 \end{cases}$, com $t \in [0,5]$

Logo, a aproximação é:

$\begin{cases} y(t+h) & = & y(t)+ f(y(t+h), t+h)h \\ & = & y(t) - 10y(t+h)h \end{cases} \rightarrow y(t+h) = \frac{y(t)}{(1 + 10h)}$

```python
def euler_impl_analitico(y_h, y0, h, t_range):
    
    """
    Resolve o problema de valor inicial com o método de Euler implícito, dada a função y(t+h).
    
    :param y_h: função de y(t+h)
    :param y0: ponto inicial do problema
    :param h: passo escolhido
    :param t_range: intervalo em t ([lower, upper])
    
    :returns (t,y): estimação de y nos valores em t
    """
        
    y_ant = y0[1]
    t_ant = y0[0]
    
    t_est = [t_ant]
    y_est = [y_ant]
    
    while t_ant < t_range[1]:
        
        # aproximação com a função y(t+h)
        y = y_h(h, y_ant)
        
        y_est.append(y)
        t_est.append(t_ant+h)
        
        # atualização dos termos
        y_ant = y
        t_ant = t_ant + h
    
    return t_est, y_est
```

### [Método de Heun](https://en.wikipedia.org/wiki/Heun%27s_method)
*(não tem na Wikipédia pt ainda!)*

É da família a qual pertence o método de Euler, dos [métodos de Runge–Kuttauma](https://pt.wikipedia.org/wiki/Lista_de_m%C3%A9todos_Runge-Kutta). Mais especificamente, o método de Euler é um Runge-Kuttama com 1 estágio, enquanto o método de Heun é um Runge-Kuttama com 2 estágios.

Resolvemos equações ordinárias da forma:

$\begin{cases} y′ = f(t, y) \\ y(t_{0}) = y_0 \end{cases}$

Através do seguinte procedimento:

$\begin{cases}Y* = Y(t) + hf(t,Y) \\ Y(t+h) = Y + {\frac {h}{2}} \left[ f(t,Y)+f(t+h,Y*) \right] \end{cases}$ (chamamos $Y*$ de valor intermediário)

#### Exercício 3:

Considere o método de Heun também chamado de método de Euler estendido.

a) Implemente este método.

```python
def heun_method(f, y0, h, t_range):
    
    """
    Resolve o problema de valor inicial com o método de Heun (Runge-Kuttama com 2 estágios).
    
    :param f: função lambda(t,y) de y' do problema
    :param y0: ponto inicial do problema
    :param h: passo escolhido
    :param t_range: intervalo em t ([lower, upper])
    
    :returns (t,y): estimação de y nos valores em t
    """
        
    y_ant = y0[1]
    t_ant = y0[0]
    
    t_est = [t_ant]
    y_est = [y_ant]
    
    while t_ant < t_range[1]:
        
        # incremento do passo
        t = t_ant + h
        
        # valor intermediário
        y_i = y_ant + h*f(t_ant, y_ant)
        
        # iteração para a próxima aproximação de Heun
        y = y_ant + (h/2)*(f(t_ant, y_ant) + f(t, y_i))
        
        y_est.append(y)
        t_est.append(t)
        
        # atualização dos termos
        y_ant = y
        t_ant = t
    
    return t_est, y_est
```

c) Mostre que este método tem erro local de truncamento da ordem $O(h^3)$. 

*dica: use a expansão de Tayor – em duas variáveis - do método de Heun e a expansão de Taylor da
solução exata e mostre que elas coincidem até a segunda derivada. Se você quiser pode
pensar no método de Heun aplicado a EDOs autônomas, o que simplifica as expressões
mas contém o essencial da ideia)*

*R:* O [erro local de truncamento](http://butler.cc.tut.fi/~piche/numa/lecture13.pdf) é um erro introduzido em cada passo entre a solução real e a aproximação fornecida pelo método, que é um truncamento da expansão. 

No caso do método de Heun de 2 estágios, $Y(t+h)$ é obtido por:


$\begin{cases}Y* = Y(t) + hf(t,Y) \\ Y(t+h) = Y + {\frac {h}{2}} \left[ f(t,Y)+f(t+h,Y*) \right] \end{cases}$ 


Abrindo a expressão, temos:

$\begin{align}
Y(t+h) & = Y + \frac{h}{2}f(t,Y)+ \frac{h}{2}f(t+h,Y(t) + hf(t,Y)) \\
& = Y + hf(t, Y) + \frac{h^2}{2}(f_{t}(t,Y) + f(t, Y)f_{y}(t, Y)) + O(h^{3}) \end{align}$

Enquanto a expansão de Taylor truncada na 2ª derivada em $\hat{Y}(t+h)$ é:

$\begin{align}
\hat{Y}(t+h) & = \hat{Y} + h\hat{Y'} + \frac{h^2}{2}\hat{Y''} + O(h^{3}) 
\\
& = Y + hf(t, Y) + \frac{h^2}{2}(f_{t}(t,Y) + f(t, Y)f_{y}(t, Y)) + \hat{O}(h^{3}) 
\end{align}$

E podemos observar que ambas coincidem até o termo da 2ª derivada. Logo, $|\hat{Y} - Y|$ tem ordem $O(h^3)$.
